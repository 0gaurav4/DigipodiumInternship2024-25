```python
import instaloader

# Ensure the instaloader instance is created
mod = instaloader.Instaloader()

# Log in to Instagram
try:
    mod.login(insta_username, insta_password)
    print("Login successful")
except instaloader.exceptions.BadCredentialsException:
    print("Login failed: Bad credentials")
except instaloader.exceptions.ConnectionException:
    print("Login failed: Connection error")
except Exception as e:
    print(f"Login failed: {e}")

# Ensure login is successful before proceeding
if mod.context.is_logged_in:
    # Ask username and number of posts to download
    username = input("Enter username here: ")
    num_posts = int(input("Enter number of posts to download: "))

    # Use API to scrape comments of any post
    posts = instaloader.Profile.from_username(mod.context, username).get_posts()
    for i, post in enumerate(posts):
        if i >= num_posts:
            break
        print(post.caption)
        try:
            for comment in post.get_comments():
                print(comment.text)
        except instaloader.exceptions.ConnectionException as e:
            print(f"Failed to fetch comments: {e}")
        except instaloader.exceptions.LoginRequiredException:
            print("Login required. Logging in again...")
            try:
                mod.login(insta_username, insta_password)
                for comment in post.get_comments():
                    print(comment.text)
            except Exception as e:
                print(f"Failed to re-login and fetch comments: {e}")

    # Get likes, posts, comments 
    profile = instaloader.Profile.from_username(mod.context, username)
    print("Username:",profile.username)
    print("UserID:",profile.userid)
    print("Followers:",profile.followers)
    print("Followees:",profile.followees)
    print("Biography:",profile.biography, profile.external_url)
    print("Private:",profile.is_private)
    print("Verified",profile.is_verified)
    print("MediaCount:",profile.mediacount)
    print("IGTVCount:",profile.igtvcount)

#download posts

    posts = instaloader.Profile.from_username(mod.context, username).get_posts()
    for i, post in enumerate(posts):
        if i >= num_posts:
            break
        mod.download_post(post, target=f"{username}")

    print("Downloaded posts successfully")

# print comments

    posts = instaloader.Profile.from_username(mod.context, username).get_posts()
    for i, post in enumerate(posts):
        if i >= num_posts:
            break
        print(post.caption)
        try:
            for comment in post.get_comments():
                print(comment.text)
        except instaloader.exceptions.ConnectionException as e:
            print(f"Failed to fetch comments: {e}")
        except instaloader.exceptions.LoginRequiredException:
            print("Login required. Logging in again...")
            try:
                mod.login(insta_username, insta_password)
                for comment in post.get_comments():
                    print(comment.text)
            except Exception as e:
                print(f"Failed to re-login and fetch comments: {e}")

    print("Downloaded comments successfully")
 
else:
    print("Cannot proceed without successful login")
```

```python
import pywhatkit
from datetime import datetime, timedelta

# Function to send WhatsApp message
def send_whatsapp_message(phone_number, message, delay_minutes=1):
    now = datetime.now()
    send_time = now + timedelta(minutes=delay_minutes)
    hour = send_time.hour
    minute = send_time.minute

    try:
        pywhatkit.sendwhatmsg(phone_number, message, hour, minute)
        print(f"Message scheduled to be sent at {hour}:{minute}")
    except Exception as e:
        print(f"Failed to send message: {e}")

# Example usage
phone_number = "+919170884817"
message = "Testing"
delay_minutes = 1  # Delay in minutes

send_whatsapp_message(phone_number, message, delay_minutes)
```
